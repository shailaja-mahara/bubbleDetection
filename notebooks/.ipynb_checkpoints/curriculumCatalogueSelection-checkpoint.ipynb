```python
import os
import pandas as pd
```

```python
CATALOGUE_A_PATH = "../data/raw/catalogue/jwst_bubble_properties_A.txt"
INPUT_PATH = "../data/processed/metadata/catalogueMerged.txt"
OUTPUT_DIR = "../data/processed/metadata/"

SIZE_CLASSES = ["tiny", "small", "medium", "large", "giants", "edgeLarge"]

SIZE_PERCENTILES = [0.20, 0.40, 0.60, 0.80]
EDGE_PERCENTILE = 0.970

CURRICULUM_STAGES = {
    "stage00": ["edgeLarge"],
    "stage01": ["giants", "edgeLarge"],
    "stage02": ["large", "giants", "edgeLarge"],
    "stage03": ["medium", "large", "giants", "edgeLarge"],
    "stage04": ["small", "medium", "large", "giants", "edgeLarge"],
    "stage05": ["tiny", "small", "medium", "large", "giants", "edgeLarge"],
}

ACTIVE_STAGE = "stage00"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# using catalogue a to define the size of class bounds

def build_catalogue_a_radius_bins(catalogue_a):
    radius = catalogue_a["AVG_RAD_PC"]

    percentile_values = radius.quantile(SIZE_PERCENTILES + [EDGE_PERCENTILE])
    q20, q40, q60, q80 = [percentile_values[p] for p in SIZE_PERCENTILES]
    edge_threshold = percentile_values[EDGE_PERCENTILE]

    finite_edges = [q20, q40, q60, q80, edge_threshold]
    if len(set(finite_edges)) != len(finite_edges):
        raise ValueError(
            "Catalogue A radius percentile edges are not unique. "
            "Adjust the binning rule before assigning size classes."
        )

    bin_edges = [float("-inf"), q20, q40, q60, q80, edge_threshold, float("inf")]

    radius_bounds = pd.DataFrame(
        {
            "SIZE_CLASS": SIZE_CLASSES,
            "lower_bound_pc": bin_edges[:-1],
            "upper_bound_pc": bin_edges[1:],
        }
    )

    return bin_edges, radius_bounds


def assign_size_classes(df, bin_edges):
    df = df.copy()

    df["SIZE_CLASS"] = pd.cut(
        df["AVG_RAD_PC"],
        bins=bin_edges,
        labels=SIZE_CLASSES,
        right=False,
    ).astype(str)

    if df["SIZE_CLASS"].isna().any():
        raise ValueError("Some bubbles could not be assigned to a size class.")

    return df
```

```python
catalogue_a = pd.read_csv(CATALOGUE_A_PATH)

bin_edges, radius_bounds = build_catalogue_a_radius_bins(catalogue_a)
catalogue_a_labelled = assign_size_classes(catalogue_a, bin_edges)

print("Catalogue A radius bounds used for all curriculum stages:")
print(radius_bounds.to_string(index=False))
```

```python
bounds_path = os.path.join(OUTPUT_DIR, "curriculum/sizeClassBounds.csv")

# df_labelled.to_csv(stratified_path, index=False)
radius_bounds.to_csv(bounds_path, index=False)
print("Saved catalogue A size-class bounds:", bounds_path)
```

```python
active_catalogue_path = os.path.join(OUTPUT_DIR, f"{ACTIVE_STAGE}.txt")

print("Active stage:", ACTIVE_STAGE)
print("Active target classes:", CURRICULUM_STAGES[ACTIVE_STAGE])
```

```python
catalogue_merged = pd.read_csv(INPUT_PATH)
size_class_bounds = pd.read_csv(bounds_path)

size_class_bounds.head(6)
```

```python
active_size_classes = CURRICULUM_STAGES[ACTIVE_STAGE]
active_size_bounds = size_class_bounds[
    size_class_bounds["SIZE_CLASS"].isin(active_size_classes)
].copy()

if active_size_bounds.empty:
    raise ValueError(f"No size bounds found for active stage: {ACTIVE_STAGE}")

print("Active size class:", active_size_classes)
print("Active size bound:", active_size_bounds)
```

```python
def get_active_stage_bounds(active_size_classes, active_size_bounds):
    active_lower_bound = float(
        active_size_bounds.loc[active_size_bounds["SIZE_CLASS"] == active_size_classes[0], "lower_bound_pc"].iloc[0])

    active_upper_bound = float(
        active_size_bounds.loc[active_size_bounds["SIZE_CLASS"] == active_size_classes[-1], "upper_bound_pc"].iloc[0])

    return (active_lower_bound, active_upper_bound)


active_lower_bound, active_upper_bound = get_active_stage_bounds(active_size_classes, active_size_bounds)

print("Active lower bound:", active_lower_bound)
print("Active upper bound:", active_upper_bound)
```

```python
segregated_catalogue = catalogue_merged[
    (catalogue_merged["AVG_RAD_PC"] >= active_lower_bound)
    & (catalogue_merged["AVG_RAD_PC"] < active_upper_bound)
].copy()
```

```python
segregated_catalogue_path = os.path.join(OUTPUT_DIR, "segregatedCatalogueMerged.txt")

segregated_catalogue.to_csv(segregated_catalogue_path, index=False)

print("Saved segregated catalogue:", segregated_catalogue_path)
print("Selected bubbles:", len(segregated_catalogue))
```
