```python
import os
import pandas as pd
import matplotlib.pyplot as plt

from astropy.io import fits
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
from matplotlib.patches import Ellipse
from astropy.visualization import ZScaleInterval
```

```python
EDGE_PERCENTILE = 0.995
GROUP_ORDER = ["tiny", "small", "medium", "large", "giants", "edgeLarge"]

fits_path = "../data/raw/fits/jw02107-o039_t018_miri_f770w_i2d.fits"

catalogues = {
    "catalogueA": {
        "path": "../data/raw/catalogue/jwst_bubble_properties_A.txt",
        "label": "A"
    },
    "catalogueB": {
        "path": "../data/raw/catalogue/jwst_bubble_properties_B_fixed.txt",
        "label": "B"
    },
    "catalogueC": {
        "path": "../data/raw/catalogue/jwst_bubble_properties_C.txt",
        "label": "C"
    },
}

base_output_dir = "../outputs/figures"
```

```python
hdu = fits.open(fits_path)
image = hdu[1].data
wcs = WCS(hdu[1].header)

print("Image shape:", image.shape)

# Conversion constants
distance_pc = 9.77e6

cdelt1 = abs(hdu[1].header["CDELT1"])
cdelt2 = abs(hdu[1].header["CDELT2"])

pixel_scale_arcsec = ((cdelt1 + cdelt2) / 2) * 3600
pc_per_arcsec = distance_pc / 206265
pc_per_pixel = pixel_scale_arcsec * pc_per_arcsec

print("pc_per_pixel:", pc_per_pixel)
```

```python
def add_size_stratification(df, edge_percentile=0.995):

    df = df.copy()

    df["SIZE_CLASS"] = pd.qcut(
        df["AVG_RAD_PC"],
        q=5,
        labels=["tiny", "small", "medium", "large", "giants"],
        duplicates="drop"
    )

    edge_threshold = df["AVG_RAD_PC"].quantile(edge_percentile)

    # Convert to string so we can assign a new non-qcut label
    df["SIZE_CLASS"] = df["SIZE_CLASS"].astype(str)

    df.loc[
        df["AVG_RAD_PC"] >= edge_threshold,
        "SIZE_CLASS"
    ] = "edgeLarge"

    return df, edge_threshold
```

```python
def print_size_summary(df, catalogue_name, edge_threshold):
    print(" " + "#" * 80)
    print(f"Catalogue: {catalogue_name}")
    print(f"edgeLarge threshold: {edge_threshold:.2f} pc")
    print("Class counts:")
    print(df["SIZE_CLASS"].value_counts())

    for group in GROUP_ORDER:
        subset = df[df["SIZE_CLASS"] == group]

        if len(subset) == 0:
            continue

        print(f"{group}")
        print("Count:", len(subset))
        print("Min AVG_RAD_PC:", subset["AVG_RAD_PC"].min())
        print("Max AVG_RAD_PC:", subset["AVG_RAD_PC"].max())
```

```python
def save_distribution_plots(df, catalogue_output_dir, catalogue_label, edge_threshold):

    distribution_dir = os.path.join(catalogue_output_dir, "distributions")
    os.makedirs(distribution_dir, exist_ok=True)

    histogram_configs = (
        (
            "linear",
            "AVG_RAD_PC",
            "Distribution of Superbubble Average Radius"
        ),
        (
            "log",
            "AVG_RAD_PC (log scale)",
            "Log-Scale Distribution of Superbubble Average Radius"
        ),
    )

    for scale, xlabel, title in histogram_configs:
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.hist(df["AVG_RAD_PC"], bins=50)
        ax.set_xscale(scale)
        ax.set_xlabel(xlabel)
        ax.set_ylabel("Number of Bubbles")
        ax.set_title(f"Catalogue {catalogue_label}: {title}")
        ax.grid(True)

        output_path = os.path.join(
            distribution_dir,
            f"avg_radius_distribution_{scale}_{catalogue_label}.pdf"
        )

        fig.savefig(output_path, dpi=300, bbox_inches="tight")
        plt.close(fig)
        print("Saved:", output_path)

    # Category boundary plot using actual class boundaries from this catalogue
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.hist(df["AVG_RAD_PC"], bins=60)

    boundaries = []

    for group in GROUP_ORDER[:-1]:
        subset = df[df["SIZE_CLASS"] == group]
        if len(subset) > 0:
            boundaries.append(subset["AVG_RAD_PC"].max())

    boundaries.append(edge_threshold)

    for b in sorted(set(boundaries)):
        ax.axvline(b, linestyle="--")

    ax.set_xscale("log")
    ax.set_xlabel("AVG_RAD_PC (pc)")
    ax.set_ylabel("Number of Bubbles")
    ax.set_title(f"Catalogue {catalogue_label}: Heavy-Tailed Distribution of Superbubble Sizes")
    ax.grid(True)

    boundary_path = os.path.join(
        distribution_dir,
        f"avg_radius_distribution_categories_{catalogue_label}.pdf"
    )

    fig.savefig(boundary_path, dpi=300, bbox_inches="tight")
    plt.close(fig)
    print("Saved:", boundary_path)
    
```

```python
def save_stratified_overlay(df, selected_group, catalogue_output_dir, catalogue_label, vmin, vmax):

    df_filtered = df[df["SIZE_CLASS"] == selected_group]

    if len(df_filtered) == 0:
        print(f"WARNING: No bubbles found for selected_group = {selected_group}")
        return None

    avg_rad_pc_min = df_filtered["AVG_RAD_PC"].min()
    avg_rad_pc_max = df_filtered["AVG_RAD_PC"].max()

    fig, ax = plt.subplots(figsize=(16, 10))
    ax.imshow(image, origin="lower", cmap="gray", vmin=vmin, vmax=vmax)

    for _, row in df_filtered.iterrows():
        bubble_id = int(row["ID"])

        coord = SkyCoord(
            row["RA_DMS"],
            row["DEC_DMS"],
            frame="icrs"
        )
        
        x, y = wcs.world_to_pixel(coord)

        semi_major_pix = row["SEMI_MAJ_PC"] / pc_per_pixel
        semi_minor_pix = row["SEMI_MIN_PC"] / pc_per_pixel

        ellipse = Ellipse(
            (x, y),
            width=2 * semi_major_pix,
            height=2 * semi_minor_pix,
            angle=row["PA_DEG"],
            edgecolor="red",
            facecolor="none",
            linewidth=1.0,
            alpha=0.8
        )

        ax.add_patch(ellipse)

        ax.text(
            x,
            y,
            str(bubble_id),
            color="red",
            fontsize=6,
            ha="center",
            va="center"
        )

    ax.set_title(
        f"Catalogue {catalogue_label}: {selected_group} bubbles, "
        f"AVG_RAD_PC {avg_rad_pc_min}–{avg_rad_pc_max} pc"
    )
    ax.set_xlabel("X pixel")
    ax.set_ylabel("Y pixel")

    plt.tight_layout()

    overlay_path = os.path.join(
        catalogue_output_dir,
        f"overlay_{selected_group}.pdf"
    )

    fig.savefig(overlay_path, dpi=300, bbox_inches="tight")
    plt.close(fig)

    print("Saved overlay:", overlay_path)
    return overlay_path
```

```python
def save_all_stratified_overlays(df, catalogue_output_dir, catalogue_label, vmin, vmax):
    saved_paths = []
    
    for group in GROUP_ORDER:
        path = save_stratified_overlay(
            df=df,
            selected_group=group,
            catalogue_output_dir=catalogue_output_dir,
            catalogue_label=catalogue_label,
            vmin=vmin, 
            vmax=vmax
        )

        if path is not None:
            saved_paths.append(path)
    return saved_paths
```

```python
for catalogue_name, catalogue_info in catalogues.items():
    catalogue_path = catalogue_info["path"]
    catalogue_label = catalogue_info["label"]
    catalogue_output_dir = os.path.join(base_output_dir, catalogue_name)

    os.makedirs(catalogue_output_dir, exist_ok=True)

    print(" " + "#" * 80)
    print(f"Processing {catalogue_name}: {catalogue_path}")

    if not os.path.exists(catalogue_path):
        print(f"WARNING: File not found, skipping: {catalogue_path}")
        continue
        
    interval = ZScaleInterval()
    vmin, vmax = interval.get_limits(image)


    df_cat = pd.read_csv(catalogue_path)

    df_cat, edge_threshold = add_size_stratification(
        df_cat,
        edge_percentile=EDGE_PERCENTILE
    )

    print_size_summary(df_cat, catalogue_name, edge_threshold)
    

    save_distribution_plots(df=df_cat, catalogue_output_dir=catalogue_output_dir, catalogue_label=catalogue_label, edge_threshold=edge_threshold)

    save_all_stratified_overlays(df=df_cat, catalogue_output_dir=catalogue_output_dir, catalogue_label=catalogue_label, vmin=vmin, vmax=vmax)
    
```
