```python
from astropy.io import fits
from astropy.wcs import WCS
import numpy as np
```

```python
# Open original FITS
hdul = fits.open("../data/raw/fits/jw02107-c1007_t007_miri_f1130w_i2d_NGC1566.fits")

# Select science extension
data = hdul[1].data
header = hdul[1].header

# Crop region
y_min, y_max = 0, 2070
x_min, x_max = 0, 1750

cropped_data = data[y_min:y_max, x_min:x_max]

# Update WCS reference pixel
header["CRPIX1"] -= x_min
header["CRPIX2"] -= y_min

# Save new FITS
output_path = "../data/raw/fits/jw02107-c1007_t007_miri_f1130w_i2d_NGC1566_cropped.fits"

hdu = fits.PrimaryHDU(data=cropped_data, header=header)
hdu.writeto(output_path, overwrite=True)

print("Saved cropped FITS:")
print(output_path)
print("Shape:", cropped_data.shape)
```
