```python
import pandas as pd
import os
```

```python
CATALOGUE_PATHS = {
    "A": "../data/raw/catalogue/jwst_bubble_properties_A.txt",
    "B": "../data/raw/catalogue/jwst_bubble_properties_B_fixed.txt",
    "C": "../data/raw/catalogue/jwst_bubble_properties_C.txt",
}

OUTPUT_PATH = "../data/processed/metadata"

os.makedirs(OUTPUT_PATH, exist_ok=True)
```

```python
all_catalogues = []

for catalogue_label, path in CATALOGUE_PATHS.items():
    df = pd.read_csv(path)
    df["CATALOGUE"] = catalogue_label
    all_catalogues.append(df)

df_all = pd.concat(all_catalogues, ignore_index=True)

# Creating a global unique ID
df_all.insert(0, "GLOBAL_ID", range(1, len(df_all) + 1))

df_all.head()
```

```python
# df_reduced = df_all.drop('ARM', axis=1)

df_reduced = df_all.drop(['ARM', 'DIST_ARM_PC', 'GAL_RAD_KPC'], axis=1)

df_reduced.head()
```

```python
print("Total bubbles in mergedCatalogue.txt:", len(df_reduced))
df_reduced.head()
```

```python
merged_output_path = os.path.join(OUTPUT_PATH, "mergedCatalogue.txt")

df_all.to_csv(merged_output_path, index=False)

print("Saved full collective catalogue to", merged_output_path)
```
